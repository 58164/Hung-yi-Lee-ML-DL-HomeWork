```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\MachineLearning\Hung-yi-Lee-ML-DL-HomeWork\HW1\ml2021spring-hw1\covid_train.csv")

# 获取最后一列的列名
target_col = df.columns[-1]

# 选择除 id 和 target 之外的所有列作为特征
feature_cols = df.columns[1:-1]   # 从第2列到倒数第2列

# 计算特征与目标的相关性
correlations = df[feature_cols.tolist() + [target_col]].corr()[target_col].drop(target_col).sort_values(ascending=False)

# 筛选出绝对值不小于 0.4 的相关系数
strong_corr = correlations[abs(correlations) >= 0.4]

print(strong_corr)
```
